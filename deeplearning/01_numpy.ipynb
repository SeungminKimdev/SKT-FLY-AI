## python 라이브러리 속도 비교 예제

```python
import random

rand_arr = [random.random() for i in range(100_000)] # python - list
type(rand_arr)
```

```python
%%timeit
# 이 셀이 실행되는 시간
import statistics

statistics.mean(rand_arr)
```

```python
np_rand_arr = np.array(rand_arr)
type(np_rand_arr)
```

```python
%%timeit
import statistics

np.mean(np_rand_arr)
```

## Boolean 인덱싱

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([True,False,True,True,True])

arr1[arr2] # arr2의 True 자리의 데이터만 가져옴
```

```python
arr1[arr1 > 2] # 조건문도 가능
```

## Fancy 인덱싱

```python
arr = np.array([1,2,3,4,5])
idx = np.array([1, 3])

arr[idx] # 1, 3 자리의 데이터를 가져옴. index로 가져오는 기능
```

```python
arr = np.array([1,2,3,4,5])
idx = np.array([[1, 3],[0,1]])

arr[idx] # 다차원도 가능. idx를 2차원으로 설정하면 2차원으로 만들어서 가져옴
```

```python
arr = np.array([1,2,3,4,5])
arr[0] = -1
arr # array의 값이 바뀌는 것이 아니라 새로운 object를 만드는 것 -> 자원 소모 많음(리스트와 차이점). ex) loop시 리스트로 진행 후 np.array로 바꿈
```

## 형태 변환

```python
arr = np.array([1,2,3,4,5,6])
arr.reshape(2,3)
```

```python
arr.reshape(-1) # 1차원 배열로 펴기, -1이 알아서 채워준다고 생각하면 됨
#arr.flatten()
#arr.ravel()
```

```python
arr1 = np.array([1,2,3,4,5,6])
arr2 = arr1.reshape(2,3)
arr2
```

```python
np.sum(arr2, axis=1) # axis= 축에 따라서 계산, np.mean도 동일
```
