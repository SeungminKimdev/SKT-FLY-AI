# Seaborn

## Import

```python
import seaborn as sns

sns.__version__
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips, x='tip')
```

```python
sns.barplot(data=tips, x='day', y='tip', hue='time', errorbar=None)
```

```python
sns.boxplot(data=tips, y='tip', hue='time', width=0.4, gap=0.4)
```

```python
sns.violinplot(data=tips, x='day', y='tip', hue='smoker', gap = 0.4)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(10,4))
sns.boxplot(data=tips, y='tip', hue='time', width=0.4, gap=0.4, ax=axes[0])
sns.violinplot(data=tips, x='day', y='tip', hue='smoker', gap = 0.4, ax=axes[1])
plt.show()
```
