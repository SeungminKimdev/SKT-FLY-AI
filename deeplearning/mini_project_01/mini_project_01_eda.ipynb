광고 클릭과 성별사이에 큰 상관관계는 없었고, 40대 중후반의 나이를 가진 사람들의 광고 클릭이 많은 것으로 나타났다.

```python
plt.figure(figsize=(12, 6))
sns.violinplot(hue='clicked_on_ad', x='gender', y='area_income', data=data, split=True)
plt.title('Area Income Distribution by Gender and Click on Ad Status')
plt.show()
#남자같은 경우, 소득에 따른 클릭 여부는 차이 없음.
#그러나 여자의 경우, 소득이 낮을수록 광고를 많이 클릭하는 경향이 있고 소득이 높을수록 광고를 클릭하지 않는 경향이 있음.
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
sns.violinplot(data=data, y='daily_time_spent_on_site', hue='clicked_on_ad', gap = 0.4, ax=axes[0])
axes[0].set_title('daily time spent')
sns.violinplot(data=data, y='daily_internet_usage', hue='clicked_on_ad', gap = 0.4, ax=axes[1])
axes[1].set_title('daily internet usage')
plt.show()
# 시간별 사이트 이용량과 일별 인터넷 사용량 모두 광고 클릭 여부에 차이가 없음
```

```python
topic_list = data['ad_topic_line'].value_counts()
top_topic_list = list(topic_list.index[:10])
data['ad_topic_line'].value_counts().head(10)
```

```python
fig, axes = plt.subplots(1,1,figsize=(8,8))
sns.histplot(data[data['ad_topic_line'].isin(top_topic_list)], y='ad_topic_line', hue='clicked_on_ad', multiple="dodge", shrink=.8)
plt.title('Clicked on ad by top topic line')
plt.show()
# 토픽이 많은 10개의 제목별 클릭 수 비교
# 특정 광고 제목에서 광고 클릭 빈도가 많은 것을 볼 수 있음
```

```python
topic_list = data['country'].value_counts()
top_topic_list = list(topic_list.index[:10])
data['country'].value_counts().head(10)
```

```python
fig, axes = plt.subplots(1,1,figsize=(8,8))
sns.histplot(data[data['country'].isin(top_topic_list)], y='country', hue='clicked_on_ad', multiple="dodge", shrink=.8)
plt.title('Clicked on ad by County')
plt.show()
# 인구 상위 10개 국가의 광고 클릭 수를 비교
# 특정 나라에서 광고 클릭 빈도가 높은 것을 확인
```
