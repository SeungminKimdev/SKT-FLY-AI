```python
sns.scatterplot(data=df, x='daily_time_spent_on_site', y='daily_internet_usage')
plt.title('spent_on_site vs daily_internet_usage')
plt.show()
# 일일 사이트 이용 시간 vs 일일 인터넷 사용량
# 두 컬럼간 상관관계가 없음
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
sns.histplot(data=df, x='daily_time_spent_on_site', ax=axes[0])
axes[0].set_title('daily time spent')
sns.histplot(data=df, x='daily_internet_usage', ax=axes[1])
axes[1].set_title('daily internet usage')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
sns.violinplot(data=df, x='daily_time_spent_on_site', hue='clicked_on_ad', gap = 0.4, ax=axes[0])
axes[0].set_title('daily time spent')
sns.violinplot(data=df, x='daily_internet_usage', hue='clicked_on_ad', gap = 0.4, ax=axes[1])
axes[1].set_title('daily internet usage')
plt.show()
# 시간별 사이트 이용량과 일별 인터넷 사용량 모두 광고 클릭 여부에 차이가 없음
```

```python
df['ad_topic_line'].value_counts().head(10)
```

```python
topic_list = df['ad_topic_line'].value_counts()
top_topic_list = list(topic_list.index[:10])
top_topic_list
```

```python
fig, axes = plt.subplots(1,1,figsize=(8,8))
sns.histplot(df[df['ad_topic_line'].isin(top_topic_list)], y='ad_topic_line', hue='clicked_on_ad', multiple="dodge", shrink=.8)
plt.title('Clicked on ad by top topic line')
plt.show()
# 토픽이 많은 10개의 제목별 클릭 수 비교
# 특정 광고 제목에서 광고 클릭 빈도가 많은 것을 볼 수 있음
```

```python
df['country'].value_counts()
```

```python
topic_list = df['country'].value_counts()
top_topic_list = list(topic_list.index[:10])
top_topic_list
```

```python
fig, axes = plt.subplots(1,1,figsize=(8,8))
sns.histplot(df[df['country'].isin(top_topic_list)], y='country', hue='clicked_on_ad', multiple="dodge", shrink=.8)
plt.title('Clicked on ad by County')
plt.show()
# 인구 상위 10개 국가의 광고 클릭 수를 비교
# 특정 나라에서 광고 클릭 빈도가 높은 것을 확인
```
