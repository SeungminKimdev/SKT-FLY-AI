# Matplotlib

## Import

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

import matplotlib
matplotlib.__version__
```

## Plot

```python
#x = np.array([1, 2, 3, 4, 5, 6])
x = np.arange(0,30,0.1)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(10,4))
plt.plot(x, y, color='r', label='sin')
plt.plot(x, z, color='b', label='cos')
plt.title('sin plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
x = np.arange(0,30,0.1)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots()
fig.set_size_inches(8,4)
ax.plot(x, y, color='r', label='sin')
ax.plot(x, z, color='b', label='cos')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()
plt.show()
```

## 산점도

```python
x = np.random.rand(30)
y = np.random.rand(30)

plt.scatter(x, y, marker='*', color='r', s=150)
plt.show()
```

## 히스토그램

```python
x = np.random.randn(1000)

plt.hist(x, bins=20)
plt.show()
```

## Bar plot

```python
x = np.array(['A','B','AB','O'])
y = np.array([5, 10, 3, 7])

plt.bar(x,y)
plt.show()
```

## Box plot

```python
D = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))

plt.boxplot(D)
plt.show()
```

## 여러개 plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
w = np.tan(x)
```

```python
plt.figure(figsize=(6,6))

plt.subplot(2, 2, 1)
plt.plot(x, y, color='r')
plt.subplot(2, 2, 2)
plt.plot(x, z, color='b')
plt.subplot(2, 2, 3)
plt.plot(x, w, color='g')
plt.subplot(2, 2, 4)
plt.plot(x, x, color='k')

plt.show()
```

```python
fig, axes = plt.subplots(2, 2)

axes[0][0].plot(x, y, color='r')
axes[0][1].plot(x, z, color='g')
axes[1][0].plot(x, w, color='b')
axes[1][1].plot(x, x, color='k')

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8,4))
axes = axes.flatten()

for i in range(4):
    axes[i].plot(x, y, color='r')

plt.show()
```
